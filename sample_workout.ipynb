```python
import requests
from bs4 import BeautifulSoup

# URL of the website
url = "https://gem.gov.in/cppp"

# Send a GET request to the website
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find tender details (adjust selectors based on the actual HTML structure)
    tenders = soup.find_all('div', class_='tender-item')  # Example class name
    
    for tender in tenders:
        title = tender.find('th', class_='Title/Ref.No./Tender Id').text.strip()  # Example tag and class
        start_date = tender.find('th', class_='Tender Opening Date').text.strip()  # Example tag and class
        closing_date = tender.find('th', class_='Bid Submission Closing Date').text.strip()  # Example tag and class
        scope = tender.find('th', class_='Organisation Name').text.strip()  # Example tag and class
        
        print(f"Title: {title}")
        print(f"Start Date: {start_date}")
        print(f"Bid Closing Date: {closing_date}")
        print(f"Scope: {scope}")
        print("-" * 40)
else:
    print("Failed to fetch the webpage. Please check the URL or your internet connection.")
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the website
url = "https://gem.gov.in/cppp"  

# Send a GET request to the website
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Locate the table (adjust selectors based on the actual HTML structure)
    table = soup.find('table', class_='table')  # Locate the table with the correct tag and class
    
    # Extract table headers
    headers = [header.text.strip() for header in table.find('thead').find_all('th')]
    
    # Extract table rows
    rows = []
    for row in table.find_all('tr')[1:]:  # Skip the header row
        cells = row.find_all('td')
        if len(cells) == len(headers):  # Ensure the row matches the headers
            row_data = {headers[i]: cells[i].text.strip() for i in range(len(headers))}
            rows.append(row_data)
    
    # Print the list of dictionaries
    print(rows)
else:
    print("Failed to fetch the webpage. Please check the URL or your internet connection.")
```
